## Fixtures

## Fixture stats
For now, we will focus on Italian Serie A Championship.

```python
for league in [x for x in leagues['leagues'] if x['country_code'] == 'IT' and x['name'] == 'Serie A' and x['is_current'] == 0][0:1]:
    fixtures = get_fixtures(league['league_id'])
    
    # Download fixture-level statistics
    path_dir = ('-'.join([league['league_id'], league['name'], league['country_code'], str(league['season'])])).replace(' ', '_')
    path_dir = os.path.join('data', 'fixture_stats', path_dir)
    print(path_dir)
    if not os.path.isdir(path_dir):
        os.mkdir(path_dir)
    for fixture in fixtures['fixtures']:
        fixture_id = fixture['fixture_id']
        print(fixture_id)
        if not os.path.exists(os.path.join(path_dir, f'fixture_stats_{fixture_id}.json')):
            fixture_stats = get_fixture_stats(fixture_id, path_dir)
            time.sleep(10)
            
    # Download player-levl statistics
    path_dir = ('-'.join([league['league_id'], league['name'], league['country_code'], str(league['season'])])).replace(' ', '_')
    path_dir = os.path.join('data', 'player_stats', path_dir)
    print(path_dir)
    if not os.path.isdir(path_dir):
        os.mkdir(path_dir)
    for fixture in fixtures['fixtures']:
        fixture_id = fixture['fixture_id']
        print(fixture_id)
        if not os.path.exists(os.path.join(path_dir, f'player_stats_{fixture_id}.json')):
            player_stats = get_player_stats(fixture_id, path_dir)
            time.sleep(10)
```
